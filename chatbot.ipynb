```python
# Step 1: Install the Google Generative AI SDK
!pip install -q google-generativeai

# Step 2: Import libraries
import google.generativeai as genai

# Step 3: Configure with your API key
genai.configure(api_key="AIzaSyCSj6SkuND0jQaa2DDwFRTnPJoqtSOnZrQ")

# Step 4: Use a supported model
model = genai.GenerativeModel("gemini-flash-latest")  # or "gemini-pro-latest"

# Step 5: Start a chat session
chat = model.start_chat(history=[])

# Step 6: Define chatbot function
def chatbot(user_input):
    response = chat.send_message(user_input)
    return response.text

# Step 7: Run chatbot loop
print("Chatbot ready! Type 'exit' to quit.\n")
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break
    reply = chatbot(user_input)
    print("Chatbot:", reply)
```
